```python
data = pd.read_csv("Iris.csv")

X = data.drop(columns=["Id", "Species"]).values
y = data.["species"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random state=123)
#X, y: Features and target to split
#test_size=0.2: 20% of data goes to testing, 80% to training
#random_state=123: Seed for reproducible random splitting
```

1. Decision Node (internal node)
Used when the tree needs to split the data further. Contains:

feature_idx: Which feature to split on (e.g., column 2 = petal length)

threshold: The value to split at (e.g., petal length > 2.5 cm)

info_gain: How much this split reduces uncertainty (e.g., Gini gain or entropy reduction)

left: Reference to the left child node (for samples meeting the condition)

right: Reference to the right child node (for samples not meeting the condition

```python
class Node:
    def __init__(self, feature_idx=None, threshold=None, info_gain=None, 
                 left=None, right=None, value=None):
        # Decision Node
        self.feature_idx = feature_idx
        self.threshold = threshold
        self.info_gain = info_gain
        self.left = left
        self.right = right
        
        # Leaf Node
        self.value = value
```
